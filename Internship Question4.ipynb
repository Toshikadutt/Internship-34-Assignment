```python
# import required libraries
from bs4 import BeautifulSoup
import requests
```

# Send get request to the webpage server to get the source code of the page

```python
page=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

Scraping multiple president names

```python
# Now we have all the tags
# Now we willthe text from these tags one by one by Looping over these tags

president_name = []  # empty list for store
for i in soup.find_all('div',class_="presidentListing"):
    president_name.append(i.text.split('\n')[1].split('(')[0])   
president_name
```

Scraping multiple terms

```python
president_name = []  # empty list for store
for i in soup.find_all('div',class_="presidentListing"):
    president_name.append(i.text.split('\n')[2])
    
president_name
```
